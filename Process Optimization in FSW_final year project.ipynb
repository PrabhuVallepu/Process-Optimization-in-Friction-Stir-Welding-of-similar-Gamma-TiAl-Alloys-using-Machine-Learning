```python
def random_sample(min_value,max_value,num_samples):
  feature_data = np.random.uniform(min_value,max_value,num_samples)
  feature_data = [int(i) for i in feature_data]
  return feature_data
```

```python
rotational_speed = random_sample(2000,2500,50)
Friction_time = random_sample(30,50,50)
Upset_pressure = random_sample(300,460,50)
Friction_pressure = random_sample(130,170,50)
Tilt_Angle = random_sample(1,3,50)
```
